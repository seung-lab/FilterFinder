```python
import os,sys,inspect
currentdir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parentdir = os.path.dirname(currentdir)
sys.path.insert(0,parentdir) 

import cv2
import numpy as np
import matplotlib.pyplot as plt
import time
import h5py
from PIL import Image
from scipy.misc import imresize
import src.visual as vis

path = '/FilterFinder/data/Dodam/'
path_s = '/FilterFinder/data/prealigned/'
metadir = '/FilterFinder/data/prealigned/registry.txt'
```

```python
#Read data
def read(name, (x,y), size):
    with h5py.File(path_s+name+'.h5', 'r') as hf:
        data = hf.get('img')
        sample = np.array(data[x:x+size, y:y+size])
    return imresize(sample, (512, 512))/255.0

def read_processed(name, (x,y), size):
    with h5py.File(path+name+'.h5', 'r') as hf:
        data = hf.get('img')
        sample = np.array(data[x:x+size, y:y+size,0])
    return imresize(sample, (512, 512))/255.0

def getMetadata():
    import csv
    f = open(metadir, 'rt')
    metadata = list(csv.reader(f, delimiter='\t'))
    return np.array(metadata)

def getDataSample(i, x, y, shape, name, path, metadata, delta=(0,0)):
    #Check if not out of borders
   
    start = np.array([x+delta[0]-shape[0]/2, y+delta[1]-shape[0]/2])
    if(start[0]<0):
        start[0]=0
    if(start[1]<0):
        start[1]=0

    #Check if not exceeding the borders
    with h5py.File(path+name+'.h5', 'r') as hf:
        data = hf.get('img')
        sample = np.array(data[start[0]:start[0]+shape[0], start[1]:start[1]+shape[1]])
    return sample

def normalize(image):
    if len(image.shape)>2:
        for i in range(image.shape[2]):
            image[:,:,i] += abs(image[:,:,i].min())
            image[:,:,i] /= abs(image[:,:,i].max())
            image[:,:,i] *= 255.0
    else:
        image += abs(image.min())
        image /= abs(image.max())
        image *= 255.0

    return image
metadata = getMetadata()
```

```python

def get_pair((i,x,y), size, original=True):
    if original:
        path_temp = path_s
        j = 0
    else:
        path_temp = path
        j = 5000
    name1 = '1,'+str(i+j+2)+'_prealigned'
    name2 = '1,'+str(i+j+3)+'_prealigned'
    
    frame1 = getDataSample(i, 15000,15000, (512,512), name1, path_temp, metadata)
    delta = [-int(metadata[i+1,3])+int(metadata[i,3]), -int(metadata[i+1,2])+int(metadata[i,2])]
    frame2 = getDataSample(i+1, 15000,15000, (512,512), name2, path_temp, metadata, delta)
    return frame1, frame2
```

```python
frame_1, frame_2 = get_pair((30,15000,15000), (512,512), False)
vis.show(frame_1)
vis.show(frame_2)
```

```python
def show_optical_flow(frame1, frame2):
    hsv = np.zeros(shape=(1,512,512,3), dtype=np.uint8)
    hsv[:,:,:,1] = 255
    z = 0
    flow = cv2.calcOpticalFlowFarneback(frame1, frame2,  0.3, 1, 20, 3, 5, 1.2, 0)
    mag, ang = cv2.cartToPolar(flow[...,0], flow[...,1])
    hsv[z,:,:,0] = ang*180./np.pi/2.
    hsv[z,:,:,2] = cv2.normalize(mag,None,0,255,cv2.NORM_MINMAX)
    hsv[z,:,:,:] = cv2.cvtColor(hsv[z,:,:,:],cv2.COLOR_HSV2BGR)

    vis.show(hsv[z,:,:,:])


show_optical_flow(frame_1, frame_2)
```
